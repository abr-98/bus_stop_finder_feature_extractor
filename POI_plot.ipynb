```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import os
%matplotlib inline
```

```python
files=list(map(lambda e:"./Grid/"+e,sorted(os.listdir("./Grid/"),key=lambda e:int(e[1:-4]))))
```

```python
fig=plt.figure(figsize=(20,20))
fig.subplots_adjust(left=0.125, bottom=0.45, right=0.9, top=0.95, wspace=0.07, hspace=0.01)
for i,f in enumerate(files):
    ax=fig.add_subplot(4,4,i+1)
    ax.imshow(np.array(Image.open(f)))
    ax.set_xticks([])
    ax.set_yticks([])
fig.suptitle('POIs',fontsize=20)
#fig.savefig('./POIs/POI_all.png',dpi=600)
```

```python
POI_codes={
    'human_made':np.array([192,192,192]),
    'natural_land':np.array([128,64,0]),
    'attraction':np.array([255,0,255]),
    #'business':np.array([255,0,0]),
    #'government':np.array([255,255,0]),
    'medical':np.array([255,128,0]),
    'park':np.array([0,255,0]),
    'education':np.array([255,255,255]),
    'water':np.array([0,0,255])
}
```

```python
df=pd.DataFrame()
```

```python
interest='water'

percentage_data=[]

fig=plt.figure(figsize=(20,20))
fig.subplots_adjust(left=0.125, bottom=0.4, right=0.9, top=0.95, wspace=0.07, hspace=0.01)
for i,f in enumerate(files):
    ax=fig.add_subplot(4,4,i+1)
    #ax.imshow(np.array(Image.open(f)))
    target=(np.array(Image.open(f))==POI_codes[interest])
    tf_map=np.logical_and(np.logical_and(target[:,:,0],target[:,:,1]),target[:,:,2])
    ax.imshow(tf_map)
    ax.set_title("{:.2f} %".format(tf_map.mean()*100))
    percentage_data.append(tf_map.mean()*100)
    ax.set_xticks([])
    ax.set_yticks([])
    
df[interest]=percentage_data

fig.suptitle('POI_'+interest,fontsize=20)
#fig.savefig('./POIs/POI_'+interest+'.png',dpi=600)
```
